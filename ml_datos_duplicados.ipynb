```python
def encontrar_duplicados(dataframe):
    """
    Función para encontrar datos duplicados en un DataFrame
    """
    # 1. BUSCAR FILAS DUPLICADAS
    print("\n. FILAS DUPLICADAS:")
    print("-" * 30)

    filas_duplicadas = dataframe[dataframe.duplicated(keep=False)]

    if len(filas_duplicadas) > 0:
        print(f"Se encontraron {len(filas_duplicadas)} filas duplicadas")
        print("\nEjemplo de filas duplicadas:")
        print(filas_duplicadas.head())
    else:
        print("No hay filas duplicadas")

    # 2. BUSCAR VALORES DUPLICADOS EN CADA COLUMNA
    print("\n DATOS DUPLICADOS EN LAS COLUMNAS:")
    print("-" * 30)

    for columna in dataframe.columns:
        # Contar cuántos valores hay en total y cuántos son únicos
        total_valores = len(dataframe[columna])
        valores_unicos = dataframe[columna].nunique()
        duplicados = total_valores - valores_unicos

        if duplicados > 0:
            print(f"{columna}: {duplicados} valores repetidos")

    # 3. VERIFICAR SI HAY COLUMNAS IGUALES
    print("\n COLUMNAS REPETIDAS:")
    print("-" * 30)

    columnas_iguales = []

    # Comparar cada columna con las demás
    for i in range(len(dataframe.columns)):
        for j in range(i + 1, len(dataframe.columns)):
            col1 = dataframe.columns[i]
            col2 = dataframe.columns[j]

            # Verificar si las dos columnas tienen los mismos valores
            if list(dataframe[col1]) == list(dataframe[col2]):
                columnas_iguales.append(f"{col1} y {col2}")

    if len(columnas_iguales) > 0:
        print("Columnas que son iguales:")
        for par in columnas_iguales:
            print(f"  • {par}")
    else:
        print("No hay columnas repetidas")

    return filas_duplicadas

# Usar la función
resultado = encontrar_duplicados(dataframe)
```

```python
def solo_duplicados(dataframe, limite=30):

    print(f"\n FILAS COMPLETAS DUPLICADAS (mostrando {limite})")
    print("-" * 30)

    filas_duplicadas = dataframe[dataframe.duplicated(keep=False)]

    if len(filas_duplicadas) > 0:
        print(filas_duplicadas.head(limite))
        if len(filas_duplicadas) > limite:
            print(f"... y {len(filas_duplicadas) - limite} filas más")
        print(f"\nTotal de filas duplicadas: {len(filas_duplicadas)}")
    else:
        print("No hay filas duplicadas")

    print(f"\n VALORES DUPLICADOS POR COLUMNA (mostrando {limite})")
    print("-" * 30)

    for columna in dataframe.columns:
        print(f"\n{columna}")
        print("   " + "-" * 20)

        valores_duplicados = dataframe[columna][dataframe[columna].duplicated(keep=False)]

        if len(valores_duplicados) > 0:

            conteo = valores_duplicados.value_counts()

            print(f"   Total valores duplicados: {len(valores_duplicados)}")
            print(f"   Valores diferentes que se repiten: {len(conteo)}")
            print(f"\n  {limite} valores que más se repiten:")
            print("   " + "-" * 10)

            for valor, cantidad in conteo.head(limite).items():
                print(f"   • {valor}: {cantidad} veces")

            if len(conteo) > limite:
                print(f"   ... y {len(conteo) - limite} valores más")
        else:
            print("No hay valores duplicados")

solo_duplicados(dataframe)
```
